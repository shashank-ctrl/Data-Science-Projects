```python
#Define X and Y
X = health.drop('Outcome',axis=1)
print(X.head())

y = health['Outcome']
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =  KNeighborsClassifier(n_neighbors=25)
```

```python
knn.fit(X_train,y_train)
```

```python
knn_prd = knn.predict(X_test)
knn_prd
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy Score of KNN Model:")
acc = accuracy_score(knn_prd,y_test)
cls = classification_report(knn_prd,y_test)
print(acc)
print(cls)
```

```python
#Preparing ROC Curve (Receiver Operating Characteristics Curve)
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# predict probabilities
probs = knn.predict_proba(X)

# keep probabilities for the positive outcome only
probs = probs[:, 1]

# calculate AUC
auc = roc_auc_score(y, probs)
print('AUC: %.3f' % auc)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positivr Rate')

# calculate roc curve
fpr, tpr, thresholds = roc_curve(y, probs)
# plot no skill

plt.plot([0, 1], [0, 1], linestyle='--')

# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
rg_prd = reg.predict(X_test)
rg_prd
```

```python
from sklearn.metrics import accuracy_score,classification_report
print("Accuracy Score of Logistic Model:")
acc = accuracy_score(rg_prd,y_test)
cls = classification_report(rg_prd,y_test)
print(acc)
print(cls)
```

```python
probs = reg.predict_proba(X)
probs = probs[:,1]

auc = roc_auc_score(y,probs)
print('AUC: %.3f' % auc)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positivr Rate')

fpr,tpr,thresholds = roc_curve(y,probs)
plt.plot([0, 1], [0, 1], linestyle='--')

plt.plot(fpr, tpr, marker='.')
```
